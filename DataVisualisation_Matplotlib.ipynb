```python
x_val = np.arange(-3,3.1,0.1)
y_val = x_val**2
plt.plot(x_val,y_val,"r*")
plt.title("Square-Function")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.xticks(np.arange(-3,3,0.7),rotation=90)
plt.show()
```

```python
x_val = np.arange(-3,3.1,0.1)
y_val = x_val**2
plt.plot(x_val,y_val,"r*")
plt.title("Square-Function")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.xlim(-10,10)     # for showing the data for a particular value
plt.show()
```

```python
x_val = np.arange(-3,3.1,0.1)
y_1 = x_val**2
y_2 = x_val ** 3

plt.plot(x_val,y_1,label = "square")
plt.plot(x_val,y_2,label = "cube")
plt.legend(["square","cube"],loc= "lower right",ncol = 2,frameon = False)
plt.show()
```

```python
x_val = np.arange(-3,3.1,0.1)
y_1 = x_val**2
y_2 = x_val ** 3

plt.plot(x_val,y_1,label = "square")
plt.plot(x_val,y_2,label = "cube")
plt.legend(["square","cube"],loc= "lower right",ncol = 2,frameon = False)
plt.text(0,0,"(0,0)")
plt.text(1,1,"(1,1)")
plt.grid()
plt.show()
```

```python
x = np.arange(20)
y1 = np.linspace(1,20,20) + np.random.rand(20)*5
y2 = np.linspace(1,20,20) + np.random.rand(20)*10
plt.scatter(x,y1 , s=y1*5 )
plt.scatter(x,y2 , s=y2*5 )
plt.legend(["Shop-1","Shop-2"])
plt.show()
```

```python
#sigmoid function #logistics
import math
```

```python
sigmoid = lambda x :1/(1+math.exp(-x))
x = np.arange(-100,100)
y = [sigmoid(i) for i in x]
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.log(x)
fig, ax = plt.subplots(3,1)
ax[0].plot(x,y1)
ax[1].plot(x,y2)
ax[2].plot(x,y3)
plt.show()

```

```python
x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.log(x)

plt.figure()

plt.subplot(2,3,1)
plt.plot(x,y1)

plt.subplot(2,3,3)
plt.plot(x,y2)
plt.subplot(2,3,5)
plt.plot(x,y3)
plt.show()

```
