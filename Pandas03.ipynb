```python
import numpy as np
import pandas as pd
```

```python
weather = pd.read_csv("weather.csv")
```

```python
weather.head()
```

```python
weather.shape   #wide data or rectangular data
```

```python
weather_melt = pd.melt(weather,id_vars=["year","month","element"],var_name = "day",value_name = "temp")
weather_melt
```

```python
weather_tidy = weather_melt.pivot_table(index=["year","month","day"],columns = "element",values = "temp").reset_index()
```

```python
# melt --> melt the columns into a single column, columns into a row values
#pivot table ---> converts values of a column into a seperate columns
```

```python
weather_tidy.to_csv("weather_tidy.csv",sep = ",")
```
