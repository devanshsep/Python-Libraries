```python
sns.histplot(iris["petal_length"],bins = 20)
plt.show()
```

```python
# kde-plot -- kernel density plot
```

```python
sns.kdeplot(iris["petal_length"])
plt.show()
```

```python
sns.scatterplot(x=iris["petal_length"],y = iris["petal_width"],data = iris , hue = "species")
plt.show()
```

```python
sns.jointplot(x="petal_length",y = "petal_width",data= iris )
plt.show()
```

```python
sns.jointplot(x="petal_length",y = "petal_width",data= iris,kind = "kde" )
plt.show()
```

```python
sns.jointplot(x="petal_length",y = "petal_width",data= iris,hue = "species" )
plt.show()
```

```python
sns.pairplot(data = iris, hue = "species")
plt.show()
```

```python
sns.countplot(x="species",data = iris)
plt.show()
```

```python
sns.boxplot(x="species",y = "sepal_length",data = iris, color = "r")
plt.show()
```

```python
sns.violinplot(x="species",y = "sepal_length",data = iris)
plt.show()
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr(),cmap="Blues",annot = True)
plt.show()
```
